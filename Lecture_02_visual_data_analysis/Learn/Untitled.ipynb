```python
data['Year_of_Release'] = data['Year_of_Release'].astype('int')
data['User_Count'] = data['User_Count'].astype('int')
data['Critic_Count'] = data['Critic_Count'].astype('int')
data['Critic_Score'] = data['Critic_Score'].astype('int')
```

```python
data.dtypes
```

```python
useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 
               'Global_Sales', 'Critic_Score', 'Critic_Count',
               'User_Score', 'User_Count', 'Rating'
              ]
```

```python
data[useful_cols].head()
```

```python
# график продаж видео игр в различных странах в зависимости от года
data_years = data [[x for x in data.columns if 'Sales' in x] + ['Year_of_Release']]
```

```python
data_years.head()
```

```python
grouped = data_years.groupby('Year_of_Release').sum() # DataFrame
# Значит можно .plot(..)
```

```python
def set_options():
    plt.rcParams['figure.figsize'] = 12,10
    fonts = {'size':'18','color':'black','fontname':'Arial','weight':'normal'}
    plt.xlabel('Year of release', fontdict = fonts)
    plt.ylabel('Count', fontdict = fonts)
    plt.xticks(fontsize = 18, rotation = 90)
    plt.yticks(fontsize = 18)
    plt.legend(fontsize = 11)
```

# KIND = 'LINE'

```python
grouped.plot(kind = 'line')
set_options()
```

# KIND = 'BAR'

```python
grouped.plot(kind = 'bar')
set_options()
```

# KIND = 'BARH'

```python
grouped.plot(kind = 'barh')
set_options()
plt.xlabel('Count')
plt.ylabel('Year of release')
```
